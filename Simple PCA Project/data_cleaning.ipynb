```python
###
# Atanas Delevski
# ECE 407 Project
# Breast Cancer PCA
###

import numpy as np
import pandas as pd

########## Reading in Data ###################
data = pd.read_csv('breast-cancer.csv')
df = pd.DataFrame(data)

```

```python
data_class = df['Class']                   # saving class column
deg_malig = df['deg-malig']                # temporarily removing deg_malig so that normalizing the 'range' columns is easier
df = df.drop(columns=['Class', 'menopause', 
                      'breast', 'breast-quad', 
                      'irradiat', 'node-caps',
                      'deg-malig'])        # dropping un-needed columns. 
```

```python
df.head()
```

```python
ave_val = lambda val_range : (val_range[0] + val_range[1])/2
def normalize_column(dataframe, column="", new_column=""): # this is my function to turn the df's values from string ranges
    mylist = []                                            # to float values which are needed for PCA algorithm
    for val_range in dataframe[column]:
        val_range_str = val_range.split('-')
        val_range_int = [int(val_range_str[0]), int(val_range_str[1])]
        mylist.append(ave_val(val_range_int))
    dataframe[new_column] = mylist
#     if del_old:
#         del dataframe[column]
    return dataframe
```

```python
normalize_column(df, 'Age', 'Age')             # here I am appending the new, 'float' columns as new columns
normalize_column(df, 'tumor-size', 'Size')     
normalize_column(df, 'inv-nodes', 'Nodes')    
del df['tumor-size']                           # and deleting the old, 'string' columns
del df['inv-nodes']

```
